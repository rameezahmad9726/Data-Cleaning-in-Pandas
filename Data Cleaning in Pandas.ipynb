# Data Cleaning in Pandas

```python
import pandas as pd 
```

```python
df = pd.read_csv(r'/Users/mac/desktop/csvs/customer_data.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.loc[:, 'Last_Name'] = df['Last_Name'].str.replace(r'[^a-zA-Z]', '', regex=True)
df
```

```python
df.loc[:, 'Phone_Number'] = (
    df['Phone_Number']
    .astype(str)
    .str.replace(r'\D', '', regex=True)
)
df
```

```python
df.loc[:, 'Phone_Number'] = df['Phone_Number'].apply(
    lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}" if len(x) == 10 else ''
)
df
```

```python
df.loc[:, 'Paying_Customer'] = df['Paying_Customer'].replace({
    'Yes': 'Y',
    'No': 'N'
})
df.loc[:, 'Do_Not_Contact'] = df['Do_Not_Contact'].replace({
    'Yes': 'Y',
    'No': 'N'
})
df
```

```python
df = df.fillna('')
df = df.replace('None', '')
df
```

```python
df = df[df['Do_Not_Contact'] != '']
df = df[df['Phone_Number'] != '']
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.drop("Not_Useful_Column", axis=1)
df
```

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df = df.drop(columns=['Address'])
df
```

```python
df.loc[:, ['State', 'Zip_Code']] = (
    df[['State', 'Zip_Code']]
    .fillna('')
    .replace('None', '')
    .apply(lambda col: col.str.strip())
)
df
```
